```
##### 셀레니움을 이용하여 네이버에서 코로나 현황을 추출한다 #####

from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys

# 1. C:/pytest에 저장할 파일의 이름과 검색어 설정
query_txt = input('크롤링할 키워드는 무엇입니까?: ')    # 코로나
path = 'C:/pytest/'                                     # 저장 경로 
f_name = path + query_txt + '.txt'                      # 저장할 파일 이름 
```

```
# 2. 크롬 드라이버로 웹브라우저 실행
path = "c:/temp/chromedriver_83/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("http://www.naver.com")
time.sleep(2)                              # 2초간 기다린다
```

```
# 3. 검색창의 이름을 찾아서 검색어를 입력, 전송 
element = driver.find_element_by_id("query")
element.send_keys(query_txt)
element.send_keys("\n")
```

```
# 4. 검색된 내용 중 텍스트를 추출
full_html = driver.page_source
soup = BeautifulSoup(full_html, 'html.parser')
```

```
# 5. 각 현황 출력
# 확진환자
red = soup.find('div', class_='circle red level5')
print(red)

red_txt = red.find('span', class_='txt_sort')
red_num = red.find('strong', class_='num')

red_txt = red_txt.text.strip()
red_num = red_num.text.strip()

print(red_txt)
print(red_num)
```

```
# 검사진행
orange = soup.find('div', class_='circle orange level5')
print(orange)

orange_txt = orange.find('span', class_='txt_sort')
orange_num = orange.find('strong', class_='num')

orange_txt = orange_txt.text.strip()
orange_num = orange_num.text.strip()

print(orange_txt)
print(orange_num)
```

```
# 격리해제
blue = soup.find('div', class_='circle blue level5')
print(blue)

blue_txt = blue.find('span', class_='txt_sort')
blue_num = blue.find('strong', class_='num')

blue_txt = blue_txt.text.strip()
blue_num = blue_num.text.strip()

print(blue_txt)
print(blue_num)
```

```
# 사망자 
black = soup.find('div', class_='circle black level3')
print(black)

black_txt = black.find('span', class_='txt_sort')
black_num = black.find('strong', class_='num')

black_txt = black_txt.text.strip()
black_num = black_num.text.strip()

print(black_txt)
print(black_num)
```

```
# 6. 자료를 딕셔너리로 정리 
daily_status = {red_txt:red_num, orange_txt:orange_num, blue_txt:blue_num, black_txt:black_num}
print(daily_status)
```

```
# 7. 딕셔너리 자료를 C:/pytest 폴더에 저장
with open(f_name, 'w') as f:
    for key, value in daily_status.items():
        f.write('%s: %s\n' % (key, value))
```
