```
import urllib.request as ul
import xmltodict
import json
import sys

# url 부분 구성
end_point = "http://apis.data.go.kr/1360000/AsosDalyInfoService"
description = "/getWthrDataList"
myKey = "kvsR2Q6DbDj2cXYm3x2eYB3%2BNMVC8D%2BIk%2BWF4Mo6ZqYllqpp24G7jgFUXdBy3sc8nl9ld%2Br1rJyYw8PVcbdTKQ%3D%3D"

# url 부분 연결 
url = end_point+description+"?serviceKey="+myKey+"&numOfRows=31&pageNo=1&dataCd=ASOS&dateCd=DAY&startDt=20200101&endDt=20200127&stnIds=108"

request = ul.Request(url)         # url의 데이터를 요청
response = ul.urlopen(request)    # 요청받은 데이터 열기
```

```
responseData = response.read().decode('utf-8')  # 요청받은 데이터를 읽음
xd = xmltodict.parse(responseData)              # XML형식의 데이터를 dict형식으로 변환
xdj = json.dumps(xd, ensure_ascii=False)        # dict 형식의 데이터를 json형식으로 변환
xdjd = json.loads(xdj)                          # json형식의 데이터를 dict 형식으로 로딩
```
