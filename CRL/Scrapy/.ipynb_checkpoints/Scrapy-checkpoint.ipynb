# Scrapy
>대표적인 크롤링 프레임워크

- 크롤링을 좀더 안정적으로 할 수 있음 
    - Scrapy 내부에서 다양한 안정장치가 있음 
- 크롤링을 좀더 빠르게 할 수 있음 
    - 크롤링 프로그램을 동시에 여러개 실행시켜서 많은 량의 데이터 크롤링시, 시간 단축
- 다양한 크롤링 관련 기능 
    - 크롤링한 데이터를 다양한 포멧으로 저장도 가능
    
    
    
    
#### 프레임워크
    - 함수와 코드가 미리 작성됨
    - 특정 함수들 어떻게 사용하는지 정해놓은 프로그램
    

### Scrapy 사용하기

- scrapy 디렉토리에 scrapyproject 디렉토리 생성 후 
    - scrapy startproject <프로젝트명>
        - <프로젝트명> 안에는 <동일 이름 하위 디렉토리> 존재 
        - <동일 이름 하위 디렉토리> 안에는 템플릿 생성
    - cmd에서 scrapy -h 입력시 사용법

scrapy 구조
-------------
------------------
scrapy.cfg # deploy configuration file  
ecommerce/ # project's Python module, you'll import your code from here  

    _init__.py  
    items.py # project items definition file  
    pipelines.py # project pipelines file  
    settings.py # project settings file  
    spiders/ # a directory where you'll later put your spiders  
    
        __init__.py

# 크롤러(spider) 작성
터미널에서 <프로젝트>/<동일 이름 하위 디렉토리>  

    scrapy gensspider <크롤러> <크롤링페이지주소>
    scrapy genspider gmarkget www.gmarket.co.kr

간단한 크롤링 확인  
cmd 창 <프로젝트>/<프로젝트명 하위 폴더>/

    scrapy crawl <크롤러>
    scrapy crawl gmarket
    
scrapy 연습  

    o scrapy shell 'http://corners.gmarket.co.kr/Bestsellers'
    o url html을 받은 후 다음 명령을 기다리는 상태 
        view(response) : 지금 html 창을 띄운다. (다운로드 받은 파일)
        shelp() : 도움말 표시
        exit : 나가기
    o response 사용법 이해
    
        response.css() : css selector로 데이터 가져오기 
        response.css('tag.class > tag > a::text').getall()
        response.css('tag.class > tag > a::text')[숫자].getall() 인덱스 가능
   
        response.xpath('xpath 복사본') : xpath로 데이터 가져오기
        response.xpath('xpath 복사본/text()') : text만 가져오기
        
    o 정규표현식도 가능
        response.css().re('정규표현식 문법')
    
    o  items.py
            class <project명 + Item>
            title = scrapy.Field()  한후                
                from ecommerce.items import EcommerceItem
                (in pasrse)
                item = EcommerceItem()
                item['title'] = title
                yield item 
            반복마다 items.py title 필드에 데이터가 쌓인다.

    o 다양한 데이터 포맷으로 아이템 저장하기
        - csv, xml , json 포맷
        - 터미널 환경에서, ecommerce 폴더에서 다음 명령
            scrapy crawl -o 저장할파일명 -t 저장포맷
            srcapy crawl gmarket -o gmarket.csv -t csv

    o 아이템 데이터 후처리하기
        - 일부만 추출, 데이베에스들에 저장, 특별한 포맷 등
        -  pipelines.py 이용
            (in settings.py)
            ITEM_PIPELINES = {
           'ecommerce.pipelines.EcommercePipeline': 300,
            }의 주석 해제
            숫자는 우선순위 번호 0~1000 숫자중 임의로 숫자를 부여하면됨
            (엄청나게 많은 크롤링을 동시에할 때 사용)
