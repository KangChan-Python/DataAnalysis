```python
import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
keyword = input('검생어를 입력해주세요 : ')
url = 'https://search.naver.com/search.naver?where=post&sm=tab_jum&query=' + keyword
driver.get(url)

option = driver.find_element_by_css_selector('#_search_option_btn')
option.click()

during = driver.find_element_by_css_selector('#snb > div > ul > li:nth-child(2) > a')
during.click()

time.sleep(3)
start = driver.find_element_by_css_selector('#blog_input_period_begin')
end = driver.find_element_by_css_selector('#blog_input_period_end')
start.send_keys('20200710')

end.send_keys(Keys.RETURN)
end.send_keys('20200706')
end.send_keys(Keys.RETURN)

apply = driver.find_element_by_css_selector('#_nx_option_date > div > span > button > span')
apply.click()
```

```python
num = driver.find_element_by_css_selector('#main_pack > div.blog.section._blogBase._prs_blg > div > span').text
num = int(num.split('/')[1].replace('건','').replace(',','').strip())
```

```python

page = (num // 10) +1
last_index = math.ceil((num% 100) /10)
```

```python

list1 = []
```

```python
blog = driver.find_elements_by_class_name('sh_blog_top')
```

```python
for i in range(0,10):
    blog[i].find_element_by_css_selector('dl > dt > a').click()

    driver.switch_to.window(driver.window_handles[-1])
    time.sleep(3)

    frame = driver.find_element_by_id('mainFrame')
    
    nurl = frame.get_attribute('src')

    driver.get(nurl)
    time.sleep(3)
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')


    span_tags =soup.select('span.se-fs-')


    temp = ''
    for i in span_tags:
        temp += i.text +'\n'
    print(temp)

    list1.append(temp)

    # 현재 탭 닫기
    driver.close()

    # 맨 처음 탭으로 변경(0번 탭)
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(3)
```
